```python
# Them vao thu vien
import pandas as pd
```

```python
df = pd.read_csv('/content/Walmart_Sales.csv')
```

```python
# Xem thong tin bang
df.info()
```

```python
df.head(10)
```

```python
# Kiem tra du lieu missing
df.isnull().values.any()
```

```python
# Tong hop cac du lieu missing values
missing_values = df.isnull().sum()
# Tong h∆°p cac du lieu duplicate row values
duplicate_row_values = df.duplicated().sum()
# Display missing values and duplicate row values
missing_values, duplicate_row_values
```

```python
# Check for duplicate row
duplicates = df.duplicated()
duplicate_rows = df[duplicates]
duplicate_count = duplicate_rows.shape[0]
print(f'Number of duplicate rows: {duplicate_count}')
print(duplicate_rows.head())
```

```python
# Remove missing data
df.dropna(axis=0, inplace = True)
# Remove duplicate row
df = df.drop_duplicates()
```

```python
#recheck
missing_values_sample = df.isnull().sum()

# Check for duplicate rows in the sample
duplicate_rows_sample = df.duplicated().sum()

# Display the cleaned sample data
cleaned_data_info_sample = {
    "missing_values": missing_values_sample,
    "duplicate_rows": duplicate_rows_sample
}

cleaned_data_info_sample
```

```python
# Cot Date data co dang ngay-thang-nam n√™n can doi lai thang-ngay-nam
# Tach cot Date thanh 3 cot dd_ngay, mm_thang, yyyy_nam
df[['dd','mm','yyyy']] = df['Date'].str.split('-', expand = True)
# Nhom 3 cot vua tach theo thu tu mm,dd,yyyy
df["N Date"] = df[['mm','dd','yyyy']].agg('-'.join, axis = 1)
# Xoa cac cot khong su dung
df.drop(['Date','mm','dd','yyyy'], axis = 1, inplace = True)
df.head(10)
```

```python
# Remove cot N Date va chen vao vi tri cot thu 1
col = df.pop('N Date')
df.insert(1, 'N Date', col)
```

```python
# Recheck thong tin df
df.info()

# File data dat yeu cau
```

```python
# Xuat file data da clean
df.to_csv('file_Walmart_slae_final.csv')
```
